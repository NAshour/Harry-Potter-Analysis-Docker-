```python
import pandas as pd 
```

```python
df = pd.read_csv('books.csv')
df
```

```python
df = df.drop(['goodreads_book_id', 'best_book_id', 'work_id', 'original_title', 'language_code', 'image_url', 'small_image_url'] , axis = 1)
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
hdf = df[df['title'].str.contains("harry potter", case=False)]
hdf
```

```python
sorted_hdf = hdf.sort_values(by='books_count', ascending=False)
best_selling = sorted_hdf.iloc[0]['title']
print(best_selling , ', number of sales:' , sorted_hdf.iloc[0]['books_count'])
```

```python
avg_rating = hdf['average_rating'].sum()/len(hdf)
avg_rating
```
